```python
def absoluteFilePaths(directory):
    for dirpath,_,filenames in os.walk(directory):
        for f in filenames:
            yield os.path.abspath(os.path.join(dirpath, f))


def absolute_file_paths(directory):
    return glob(join(directory, "**"))
```
